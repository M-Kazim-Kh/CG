```python
import matplotlib.pyplot as plt
import numpy as np

# Circle parameters
h = 2  # x-center
k = -3 # y-center
r = 5  # radius

theta = np.linspace(0, 2 * np.pi, 100)
# print(theta)
x = h + r * np.cos(theta)
y = k + r * np.sin(theta)

plt.plot(x, y)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Circle centered at ({}, {}) with radius {}".format(h, k, r))
plt.grid(True)
plt.show()
```
