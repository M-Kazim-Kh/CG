```python
""" Handling Keyboard & Mouse Events """

import pygame

# Initialize Pygame
pygame.init()

# Set window dimensions
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))

# Set window title
pygame.display.set_caption("Keyboard and Mouse Events")

# Define colors (RGB format)
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)
GREEN = (0, 255, 0)
BLUE = (0, 0, 255)

# Circle properties
circle_pos = (WIDTH // 2, HEIGHT // 2)  # Initial position
circle_radius = 50
circle_color = RED

# Text properties
font = pygame.font.Font(None, 36)  # Default font, size 36
text1 = "Press (R) for Red, (G) for Green, (B) for Blue"
text2 = "Press keys or click the mouse."

# Main loop
running = True
while running:
    # Handle events
    for event in pygame.event.get():
        # Quit event
        if event.type == pygame.QUIT:
            running = False

        # Keyboard events
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_ESCAPE:  # Exit if Escape key is pressed
                running = False
            elif event.key == pygame.K_r:  # Change circle color to red
                circle_color = RED
                text = "Circle color changed to RED!"
            elif event.key == pygame.K_g:  # Change circle color to green
                circle_color = GREEN
                text = "Circle color changed to GREEN!"
            elif event.key == pygame.K_b:  # Change circle color to blue
                circle_color = BLUE
                text = "Circle color changed to BLUE!"

        # Mouse events
        if event.type == pygame.MOUSEBUTTONDOWN:
            if event.button == 1:  # Left mouse button
                circle_pos = event.pos  # Move circle to mouse click position
                text = "Circle moved to mouse click position!"

    # Fill the screen with a background color (white in this case)
    screen.fill(WHITE)

    # Draw the circle
    pygame.draw.circle(screen, circle_color, circle_pos, circle_radius)

    # Render the text
    text_surface1 = font.render(text1, True, BLACK)
    text_surface2 = font.render(text2, True, BLACK)
    screen.blit(text_surface1, (10, 10))  # Display text at (10, 10)
    screen.blit(text_surface2, (10, 50))  # Display text at (10, 10)

    # Update the display
    pygame.display.flip()

# Quit Pygame
pygame.quit()
```
