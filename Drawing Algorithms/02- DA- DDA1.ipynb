```python
""" Digital Differential Analyzer (DDA) using PyGame """

import pygame
# import pygame.gfxdraw

# Function to draw a line using DDA
def DDA(x0, y0, x1, y1):
    # Calculate differences
    dx = x1 - x0
    dy = y1 - y0

    # Determine the number of steps
    steps = max(abs(dx), abs(dy))

    # Calculate increments
    x_inc = dx / steps if steps !=0 else 0
    y_inc = dy / steps if steps !=0 else 0

    # Initialize starting point
    x, y = x0, y0

    Pixels=[]

    # Create the line' Pixels
    for _ in range(steps+1):
        Pixels.append((round(x),round(y)))
        x += x_inc
        y += y_inc

    return Pixels


if __name__ == "__main__":
    # Set up display
    screen = pygame.display.set_mode((800, 600))
    pygame.display.set_caption("Digital Differential Analyzer (DDA) Algorithm")
    screen.fill((255,255,255))

    x0, y0, x1, y1 = 50, 50, 700, 500  # Line endpoints
    pixels=DDA(x0, y0, x1, y1)
    # print(pixels)

    for x,y in pixels:
        screen.set_at((round(x),round(y)), (0,0,0))

    # Update the display
    pygame.display.flip()


    # Main loop to keep the window open
    running = True
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
            elif event.type == pygame.KEYDOWN and event.key == pygame.K_ESCAPE:
                running = False

    pygame.quit()
```
